Подключим модуль pipeline

```python
from transformers import pipeline
```

Определение тональности текста. Создаем pipline для обработки текст

```python
classifier = pipeline("sentiment-analysis")
```

Теперь используем пайплайн для определения тональности

```python
results = classifier("I love natural language processing.")
results
```

```python
results = classifier("I do not love natural language processing.")
results
```

```python
results = classifier("I hate natural language processing.")
results
```

Теперь сделаем пайплайн для определения тональности для русского языка. Используем модель RuBERT for Sentiment Analysis.

```python
classifier = pipeline("sentiment-analysis", model="blanchefort/rubert-base-cased-sentiment")
```

```python
results = classifier("Я ненавижу тех, кто любит обработку естественного языка")
results
```

```python
results = classifier("Я не оправдываю тех, кто  не любит обработку естественного языка")
results
```

```python
results = classifier("Я восхищаюсь теми, кто любит обработку естественного языка")
results
```

Ответы на вопросы

```python
qa_model = pipeline("question-answering", model="AlexKay/xlm-roberta-large-qa-multilingual-finedtuned-ru")
```

```python
context="""Меня зовут Черноскутов Дмитрий, я занимаюсь нейронных сетей и изучением искусственного интеллекта.
           Сейчас я работаю в Уральском федеральном университете, где я применяю компьютерные техноллгии применимо к задачам электроэнергетики. В программу входят задачи по компьютерному зрению,
           обработке больших данных, обработки электрических сигналов, прогнозированию событий и вероятности. Также есть интересный курс по 
           применению машинного обучения под различные задачи.
         """
```

```python
question="Как называется направление, которым занимается Черноскутов Дмитрий?"
```

```python
results = qa_model(question = question, context = context)
results
```

```python
question="Что включено в научные цели, решаемые Черноскутовым Дмитрием?"
```

```python
results = qa_model(question = question, context = context)
results
```
